```python
X = df.iloc[:, [3,4]].values
```

```python
plt.figure(figsize = (15,6))
n=0
for x in ['Age','Annual Income (k$)','Spending Score (1-100)']:
    n += 1
    plt.subplot(2,3,n)
    plt.subplots_adjust(hspace=0.2,wspace = 0.2)
    sns.histplot(df[x],bins = 20)
    plt.title('DistPlot of {}'.format(x))
plt.show();
```

```python
plt.figure(figsize = (15,4))
sns.countplot(y='Genre',data = df)
plt.title('Spending Score (0-100): Genre Distribution')
plt.show();
```

```python
plt.figure(1,figsize = (15,6))
for Genre in ['Male','Female']:
    plt.scatter(x = 'Age',y = 'Annual Income (k$)',data =df[df['Genre'] == Genre],s = 200,alpha = 0.7,label = Genre)
plt.xlabel('Age'),plt.ylabel('Annual Income (k$)')
plt.title('Age vs Annual Income wrt Genre')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans 
wcss = []
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS Values')
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters = 5, init='k-means++', random_state=0)
```

```python
y_kmeans = kmeansmodel.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0,0], X[y_kmeans == 0,1], s= 100, c ="red" , label= 'Customer 1')
plt.scatter(X[y_kmeans == 1,0], X[y_kmeans == 1,1], s= 100, c ="cyan" , label= 'Customer 2')
plt.scatter(X[y_kmeans == 2,0], X[y_kmeans == 2,1], s= 100, c ="green" , label= 'Customer 3')
plt.scatter(X[y_kmeans == 3,0], X[y_kmeans == 3,1], s= 100, c ="yellow" , label= 'Customer 4')
plt.scatter(X[y_kmeans == 4,0], X[y_kmeans == 4,1], s= 100, c ="black" , label= 'Customer 5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=120 , c = 'magenta' , label ='Centroids')
plt.title('Cluster of customers')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
